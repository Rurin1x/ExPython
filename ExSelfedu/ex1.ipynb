# Математические функции и модуль math

```python
a = abs(-5) # модуль числа
a
```

```python
b = min(3, 45, 7, 5) # минимальное среди чисел
b1 = max(3, 45, 7, 5) # максимальное
b, b1
```

```python
с = pow(6, 2) # возведение в степень, аналог 6 ** 2
с
```

```python
d = round(1.5555, 2) # округление, аргумент = точности
d1 = round(1.55)
d, d1
```

```python
e = max(abs(-3), max(1, 3, 4, max(-3, -6, abs(-12) ))) # можно вкладывать друг в друга
e
```

#  Модуль math

```python
import math
```

```python
r = math.ceil(5.2) # округление до наибольшего целого
r1 = math.floor(5.7) # округление до наименьшего целого
r, r1
```

```python
f = math.factorial(6) # 1 * 2 * 3 * 4 * 5 * 6
f
```

```python
g = math.trunc(5.9) # отбрасывает дробную часть, аналог int()
g
```

```python
h = math.sqrt(36) # вычисляет квадратный корень
h
```

```python
j = math.pi # число pi
j1 = math.e # число e
j, j1
```

# Простые задачи

```python
import math
a, b = map(int, input().split())
c = math.sqrt(pow(a, 2) + pow(b, 2))
print(round(c, 2))
```

```python
import math
n, k = map(int, input().split())
print(round(math.factorial(n)/(math.factorial(n - k) * math.factorial(k))))
```
