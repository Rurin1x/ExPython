<h1>Целочисленная арифметика. Часть 2</h1>

```python
a = 15 // (16 % 7)
print(a)
b = 34 % a * 5 - 29 % 5 * 2
print(a + b)
```

```python
a = 82 // 3 ** 2 % 7
print(a)
```

```python
a = int(input())
b = int(input())
ap = b // a
ost = b % a
print(ap)
print(ost)
```

```python
a = int(input())
sm = a // 2
viz = a - sm
print(viz)
```

```python
a = int(input())
ch = a // 60
mn = a % 60
print(f'{a} мин - это {ch} час {mn} минут.')
```

```python
# получение цифр трёхзначного числа
num = int(input())
a = num % 10
b = (num % 100) // 10
c = num // 100
print(f"Сумма цифр = {a+b+c}")
print(f"Произведение цифр = {a*b*c}")
```

```python
num = int(input())
a = num % 10
b = (num % 100) // 10
c = num // 100
a, b, c = c, b, a
print(a, b, c, sep="")
print(a, c, b, sep="")
print(b, a, c, sep="")
print(b, c, a, sep="")
print(c, a, b, sep="")
print(c, b, a, sep="")
```

```python
x=int(input())
a=x//100
b=x%100
c=a%10
d=a//10
e=b%10
f=b//10
print('Цифра в позиции тысяч равна', d)
print('Цифра в позиции сотен равна', c)
print('Цифра в позиции десятков равна', f)
print('Цифра в позиции единиц равна', e)
```
